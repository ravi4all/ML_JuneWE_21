```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "names" : ["Sachin","Virat","Dhoni","Yuvraj","Rohit","Dhawan"],
    "score" : [102,120,56,60,23,10],
    "balls" : [100,89,33,40,34,20]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,6))
plt.title("Scores by each player")
plt.bar(df['names'], df['score'], width=0.9, color='r')
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,6))
plt.title("Balls Taken by each player")
plt.barh(df['names'], df['balls'], color='r')
plt.xlabel("Name")
plt.ylabel("Balls")
plt.show()
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,6))
plt.title("Scores by each player")
plt.hist(df['score'], rwidth=0.9)
plt.xlabel("Score")
plt.ylabel("Freq")
plt.show()
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,6))
plt.title("Scores by each player")
# plt.plot(df['balls'], df['score'])
# plt.plot(df['balls'], df['score'], 'bo')
plt.scatter(df['balls'], df['score'])
plt.xlabel("Balls")
plt.ylabel("Score")
plt.show()
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,10))
plt.title("Scores by each player")
plt.pie(x=df['score'], labels=df['names'], shadow=True, autopct='%1.2f%%',
       explode=(0,0.1,0,0,0,0), startangle=90, counterclock=False)
plt.show()
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,6))
plt.title("Scores by each player")
plt.boxplot(df['score'])
plt.xlabel("Score")
plt.ylabel("Freq")
plt.show()
```
