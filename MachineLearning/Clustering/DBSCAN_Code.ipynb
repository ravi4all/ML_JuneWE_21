```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('dbscan_data.xlsx')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df[0], df[1], s=10)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(df)
```

```python
df['Target'] = kmeans.labels_
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df[0], df[1], s=10, c=df['Target'])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=40, min_samples=6)
dbscan.fit(df)
```

```python
df['Target_DBSCAN'] = dbscan.labels_
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df[0], df[1], s=10, c=df['Target_DBSCAN'])
plt.show()
```
